```python
buckets = standard.keys()
buckets = list(buckets)
```

```python
import shutil

for index, bucket in enumerate(buckets) : 
    target_folder_path = os.path.join(folder_path, bucket)
    file_list = standard[bucket]
    print("target_folder_path :: ", target_folder_path)
    os.makedirs(target_folder_path, exist_ok=True)
    #해당 파일들은 해당 bucket으로
    for index_2, measurement in enumerate(file_list) :
        source_file_path = os.path.join(source_folder_path, measurement)
        source_file_path = source_file_path + ".csv"
        
        if os.path.exists(source_file_path):
            print("실제 존재 파일:: ", source_file_path)
            shutil.copy(source_file_path, target_folder_path)
        else : 
            print("존재하지 않는 파일 :: ", source_file_path)
```
