```python
#대용량 폴더 안의 파일들 
#원하는 갯수만큼 잘라서 
#다른 폴더 안에 넣기

import os
import sys
import pandas as pd
import shutil

os.chdir("./")
path = os.getcwd()
```

```python

#대용량 폴더
path_outdoor = path + '/' + 'outdoor' 

file_list = os.listdir(path_outdoor)

count = 0
forder_nm = ''
dest = ''
for x in file_list :        
    
    count = count + 1 
    #원하는 갯수만큼 자르기
    if ( (count == 1) or (count % 100 == 0)) : 
        forder_nm = 'data_' + str(count)
        dest =  path + '/' + forder_nm
        
        os.makedirs( dest, exist_ok = True )     
  
    #새로 만든 폴더로 이동
    shutil.copy(path_outdoor + '/' + x, dest)
    
   

```
