# 실행 코드

```python
for db_name in db_arr:
    uploadParam = {
        "databaseName" : db_name,
        "measurementsName" : '',
        "mode" : "save"
    }
    #메저먼트 리스트
    mList = db_client.measurement_list(db_name)
    print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>",len(mList), db_name)
    
    for idx, ms_name in enumerate(mList) :
        # meta 읽어오는 코드
        collect_read = ReadData(db_client, db_name, ms_name) 
        meta = collect_read.get_ms_meta()    
        if 'tag' in meta :
            print(idx , ",", ms_name, ": 메타에 키 이름 tag 존재함")
            meta['keyword'] = meta.pop('tag') # 메타 변경
            uploadParam["measurementsName"]  = ms_name
            #WriteData(uploadParam, meta).set_meta() 

        elif 'keyword' in meta : 
            print(idx , ",", ms_name, ": 메타에 키 이름 keyword 존재함")  
```
